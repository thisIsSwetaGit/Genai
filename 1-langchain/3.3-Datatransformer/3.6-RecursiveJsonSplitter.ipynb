```python
import json
import requests
json_data = requests.get('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.json').json()
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
json_splitter = RecursiveJsonSplitter(max_chunk_size=1000)
json_chunks = json_splitter.split_json(json_data[0])
json_chunks[:2]  # Display the first two chunks for brevity
```

```python
# splitter can also output in documents format
docs=json_splitter.create_documents(json_data)
for doc in docs[:2]:  # Display the first two documents for brevity
    print(doc)
```

```python
text=json_splitter.split_text(json_data[0])
print(text[:5])  # Display the first 500 characters of the text for brevity
```
