```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma:2b") #by default it uses llama2
)
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(
    [
        "alpha is the first letter in the Greek alphabet",
        "beta is second letter in the Greek alphabet",
    ]
)
```

```python
r1
```

```python
embeddings.aembed_query("what is first letter of greek alphabet?")
```

```python
# other models can be used as well
embeddings=OllamaEmbeddings(model="mxbai-embed-large")
text="meow  meow meow cat"
query=embeddings.embed_query(text)
query
```
